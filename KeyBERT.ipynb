```python
# 함수 정의

def key_extract(doc) :
  keyword = kw_model.extract_keywords(doc,keyphrase_ngram_range=(1,1),use_mmr = True,top_n = 5,diversity = 0.1) # keyword추출
  print(keyword)
  return keyword
```
