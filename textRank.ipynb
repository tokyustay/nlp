# tokenizer를 이용하여 명사만 추출하여 keyword 추출 진행한 코드
    textrank_keyword_extraction(text, top_n=3)에서 top_n 값을 조절하면서 추출 키워드 개수 조절 가능



```python
import networkx as nx
from konlpy.tag import Okt
from collections import defaultdict
import itertools

# Hyperparameter 'top_n'으로 추출할 키워드의 최종 개수를 조절
def textrank_keyword_extraction(text, top_n=3):
    # Initialize Korean tokenizer
    okt = Okt()

    # Extract words as nodes
    # Extracts only nouns
    nodes = okt.nouns(text)

    # Tokenize based on spacing
    # nodes = text.split()

    # Create a graph
    graph = nx.Graph()
    graph.add_nodes_from(set(nodes))

    # Add edges between nodes that co-occur within a window size of 2
    window_size = 2
    edges = [nodes[i:i+window_size] for i in range(len(nodes) - window_size + 1)]
    for window in edges:
        for u, v in itertools.combinations(window, 2):
            if not graph.has_edge(u, v):
                graph.add_edge(u, v, weight=0)
            graph[u][v]['weight'] += 1

    # Apply the TextRank algorithm
    scores = nx.pagerank(graph, weight='weight')

    # Sort nodes by TextRank score in descending order
    sorted_nodes = sorted(scores.items(), key=lambda item: item[1], reverse=True)

    # Return the top N keywords
    return [keyword for keyword, score in sorted_nodes[:top_n]]
```

```python
# Example text (in Korean)
korean_text = """테슬라가 중국에서 외국 기업 중 처음으로 데이터 안전 검사를 통과했다.
테슬라가 보유한 완전자율주행(FSD) 기술의 중국 시장 도입에 청신호가 켜졌다는 전망이 나온다.
일론 머스크 테슬라 최고경영자(CEO)는 오는 8월 8일 이 기술을 적용한 로보택시(무인택시)를 공개하겠다고 밝히는 등 자율주행 기술의 완성도에 자신감을 보여왔다.
최근 전기차 수요 둔화와 가격 인하 경쟁 등으로 실적 부진에 시달리는 테슬라가 FSD 중국 출시로 반전의 돌파구를 마련하는 것 아니냐는 전망이 나온다."""

# Extract keywords
keywords = textrank_keyword_extraction(korean_text)
print("Keywords:", keywords)
```
